## Most Common Listings

```python
#Lets see how the listings are distributed amongst the neighborhoods
neighborhood_dist = sf['neighborhood'].value_counts(normalize=True)*100

#plot the distribution

#create an "other" category for the neighborhoods with smallest percentages
threshold = 1
mask = neighborhood_dist > threshold
other = neighborhood_dist[~mask].sum()
other_count = neighborhood_dist[~mask].count()
dist = neighborhood_dist[mask]
dist["Other {} neighborhoods".format(other_count)] = other

#plot the distributions
dist.plot(kind = 'barh', figsize = (12,8))
plt.title('Percent of SF Airbnb Listings By Neighborhood', fontsize = 18)
plt.xlabel('Percent')
plt.ylabel('Neighborhood')
plt.gca().invert_yaxis()
plt.show()
```

## Are There Any Major Correlations?

```python
sf['room_type'].value_counts()
```

```python
#Map room types to an integer
dict = {
    'Shared room' : 0,
    'Private room' : 1,
    'Hotel room' : 2,
    'Entire home/apt' : 3
}

#Create a new column with integers for room types
sf['room_type_value'] = sf['room_type'].replace(dict)
```

```python
#Map neighborhoods to an integer
neighborhood_dict = {}

#Create unique list of neighborhoods
neighborhoods = np.unique(sf['neighborhood'])
i=0

#Loop through unique neighborhoods and map to integers in a dictionary
for n in neighborhoods:
    neighborhood_dict[n] = i
    i += 1
    
print(neighborhood_dict)

#Create the new column
sf['neighborhood_value'] = sf['neighborhood'].replace(neighborhood_dict)
```

```python
sns.set(style="white")

# Compute the correlation matrix
corr = sf.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(9, 7))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

### Correlation Findings

Visually, it appears that there are no correlations of major concern between the features in our dataset. However, under closer inspection, it seems there are two weak correlations that we can draw some insight from:

1. **The lower the listing ID, the higher the number of reviews.**
    1. This is most likely due to the fact that listings with lower IDs have been on the Airbnb platoform for a longer time. Thus, it has been rented and reviewed more often.
2. **Listings are currently more available(by number of nights) if they have higher minimum nights per stay requirements.**
    1. This shows it may be possible that most renters that use Airbnb are indeed looking for shorter term rentals. 
    

## Do Any Listings Violate SF Short Term Rentals Ordinance?

San Francisco's short-term rentals ordinance has several stupilations on what Airbnb rentals can or cannot do. Here are some important details:

 - Short term rentals must be 30 days or less 
 - Short term rentals may only be rented for 90 days per calendar year
     - There is no limit if host stays in home during the rental
     - i.e. Limit applies if entire house/apt is rented out
 - Short term rentals must be lived-in by the host for 265 days per calendar year
 - Does not apply to commercial bed & baths or hotels
  
Source: https://businessportal.sfgov.org/start/starter-kits/short-term-rental

```python
#sort for non-commercial listings
violators = sf[sf['calculated_host_listings_count'] <= 1].copy()

#sort for entire home/apt listings
violators = violators[violators['room_type'] == 'Entire home/apt']

#find possible violators
possible_bad_listings = violators[violators['minimum_nights'] > 30]
possible_bad_listings_sum = possible_bad_listings['minimum_nights'].value_counts().sum()

print('There are {} listings that are possibly in violation of the SF ordinance.'.format(possible_bad_listings_sum))
```

### Observations

Above, we found 99 non-commerical listings for entire homes/apartments that are renting in excess of the 30 day limit for short-term rentals. However, to verify these listings, it will be necessary answer the following questions for each listing:

1. Are the listings actually entire homes/apts or are the hosts simply listing it as such?
2. Are these listings actually listed by non-commercial hosts?
